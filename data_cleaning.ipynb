### Idea

As from the above data we can see the data is stacked-up in vertically we need to split and stack them 
horizontally in the process we remove the repeating columns

```python
# dropping gas and unit columns

cd = concated_dataframe.drop(['gas', 'unit'], axis=1)
```

```python
def split_dataframe_by_position(df, splits):
    """
    Takes a dataframe and an integer of the number of splits to create.
    Returns a list of dataframes.
    """
    dataframes = []
    index_to_split = len(df) // splits
    start = 0
    end = index_to_split
    for split in range(splits):
        temporary_df = df.iloc[start:end, :]
        dataframes.append(temporary_df)
        start += index_to_split
        end += index_to_split

    return dataframes

splits = split_dataframe_by_position(cd, 195)
print(splits[193])
```
